```python
Candy = pandas.concat([df,], axis = 1)
del Candy['competitorname']
Candy.head()
X = Candy[['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard','bar','pluribus','sugarpercent','pricepercent']]
y = Candy['winpercent']
```

```python
predictorcols = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard','bar','pluribus','sugarpercent','pricepercent']
import itertools
for variables in itertools.combinations(predictorcols, 1):
    print(variables)
```

```python
import statsmodels.api as sm
X2 = sm.add_constant(X)
est = sm.OLS(Y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
# as we can see on the R-squared is bigger than 0.5, therefor our X and Y have correlation.
# then i used AIC model to validation the X features.
# I randomly put the X features to all the combination, and check the lowest c.most_common(),for the best combination
```

```python
import itertools
AICs = {}
for k in range(1,len(predictorcols)+1):
    for variables in itertools.combinations(predictorcols, k):
        predictors  = X[list(variables)]
        predictors2 = sm.add_constant(predictors)
        est = sm.OLS(Y, predictors2)
        res = est.fit()
        AICs[variables] = res.aic
```

```python
from collections import Counter
c = Counter(AICs)
#c.most_common()
c.most_common()[::-10]
```
